```python
from matplotlib import pyplot as plt  
%matplotlib inline

folder = "../images"

fig = plt.figure(figsize=(20, 10))

# Read the images using OpenCV (OpenCV loads images in BGR format)
image1 = plt.imread(f'{folder}/count_avg_50bis.png')
image2 = plt.imread(f'{folder}/bleu_avg_50bis.png')
image3 = plt.imread(f'{folder}/rouge_avg_50bis.png')

plt.subplot(2, 3, 1)  # 2 rows, 2 columns, first position
plt.imshow(image1)  
plt.axis('off')  # Hide the axis labels
plt.title("Common bigrams top 50 users with topx bigrams as predict comment") 

plt.subplot(2, 3, 2)  # 2 rows, 2 columns, second position
plt.imshow(image2)  
plt.axis('off')  # Hide the axis labels
plt.title("Precision top 50 users with topx bigrams as predict comment") 

plt.subplot(2, 3, 3)  # 2 rows, 2 columns, third position
plt.imshow(image3) 
plt.axis('off')  # Hide the axis labels
plt.title("Recall top 50 users with topx bigrams as predict comment") 

image4 = plt.imread(f'{folder}/count_avg_200bis.png')
image5 = plt.imread(f'{folder}/bleu_avg_200bis.png')
image6 = plt.imread(f'{folder}/rouge_avg_200bis.png')

plt.subplot(2, 3, 4)  # 2 rows, 2 columns, third position
plt.imshow(image4) 
plt.axis('off')  # Hide the axis labels
plt.title("Common bigrams 200 random users with topx bigrams as predict comment") 

plt.subplot(2, 3, 5)  # 2 rows, 2 columns, third position
plt.imshow(image5) 
plt.axis('off')  # Hide the axis labels
plt.title("Precision 200 random users with topx bigrams as predict comment") 

plt.subplot(2, 3, 6)  # 2 rows, 2 columns, third position
plt.imshow(image6) 
plt.axis('off')  # Hide the axis labels
plt.title("Recall 200 random users with topx bigrams as predict comment") 

plt.tight_layout()
plt.show()

```
